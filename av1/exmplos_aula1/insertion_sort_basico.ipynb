```python
import numpy as np
import matplotlib.pyplot as plt
import time

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Função para medir o tempo de execução do Insertion Sort
def measure_time(arr):
    start_time = time.time()
    insertion_sort(arr)
    end_time = time.time()
    return end_time - start_time

# Gerar arrays de diferentes tamanhos
sizes = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
times_random = []
times_ascending = []
times_descending = []

for size in sizes:
    # Cria um array aleatório de tamanho 'size'
    arr_random = np.random.randint(0, 1000, size) #
    arr_ascending = np.sort(arr_random)
    arr_descending = np.sort(arr_random)[::-1]
    
    exec_time_random = measure_time(arr_random)
    exec_time_ascending = measure_time(arr_ascending)
    exec_time_descending = measure_time(arr_descending)
    
    times_random.append(exec_time_random)
    times_ascending.append(exec_time_ascending)
    times_descending.append(exec_time_descending)

# Plotar os resultados
plt.figure(figsize=(10, 6))
plt.plot(sizes, times_random, marker='o', label='Aleatório', color='red')
plt.plot(sizes, times_ascending, marker='o', label='Crescente', color='green')
plt.plot(sizes, times_descending, marker='o', label='Decrescente', color='blue')
plt.title('Tempo de Execução do Insertion Sort')
plt.xlabel('Tamanho do Array')
plt.ylabel('Tempo de Execução (segundos)')
plt.legend()
plt.grid()
plt.show()
```
