```python
import numpy as np
import matplotlib.pyplot as plt
import time

# ALGORITMO INSERTION SORT 
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j] # Essa linha Ã© responsÃ¡vel por mover os elementos
            j -= 1 # Decrementa o Ã­ndice para comparar com o prÃ³ximo elemento
        arr[j + 1] = key
    return arr

# FunÃ§Ã£o para medir o tempo de execuÃ§Ã£o do Insertion Sort
def measure_time(arr):
    start_time = time.time()
    insertion_sort(arr)
    end_time = time.time()
    return end_time - start_time

# Gerar arrays de diferentes tamanhos
# sizes = [10, 100, 1000, 10000, 100000, 1000000]
# melhorado abaixo
sizes = np.arange(10, 10000, 100) # inicia em 10, termina em 10000, com passo de 10


times_random = [] # Lista para armazenar os tempos de execuÃ§Ã£o para arrays aleatÃ³rios. VÃ£o ser preenchidos no loop abaixo e depois plotados
times_ascending = []
times_descending = []

for size in sizes:
    # Cria um array aleatÃ³rio, com nÃºmeros de 0 a 100000, mas gera 'size' a quantidade de nÃºmeros
    arr_random = np.random.randint(0, 1000001, size) # 
    arr_ascending = np.sort(arr_random)
    arr_descending = np.sort(arr_random)[::-1]
     # carriage return
    print(f" â–º", end="\r")
    
    
    exec_time_random = measure_time(arr_random)
    exec_time_ascending = measure_time(arr_ascending)
    exec_time_descending = measure_time(arr_descending)
    
    times_random.append(exec_time_random)
    times_ascending.append(exec_time_ascending)
    times_descending.append(exec_time_descending)
    
#print(f"Ãºltimo array ascendente\n {arr_ascending}")
#print(f"Ãºltimo array randÃ´mico\n {arr_random}")
#print(f"Ãºltimo array descendente\n {arr_descending}")
# Plotar os resultados
plt.figure(figsize=(10, 6))
plt.plot(sizes, times_random, label='AleatÃ³rio', color='red')
plt.plot(sizes, times_ascending, label='Crescente', color='green')
plt.plot(sizes, times_descending, label='Decrescente', color='blue')
plt.title('Tempo de ExecuÃ§Ã£o do Insertion Sort')
plt.xlabel('Tamanho do Array')
plt.ylabel('Tempo de ExecuÃ§Ã£o (segundos)')
plt.legend()
plt.grid()
plt.show()
```
