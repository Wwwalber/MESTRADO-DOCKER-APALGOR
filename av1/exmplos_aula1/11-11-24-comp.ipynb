```python
import numpy as np
import matplotlib.pyplot as plt

# Definir o intervalo de valores para n
n = np.linspace(1, 10, 100)  # Cria 100 pontos entre 1 e 100

# Imprime o array original
print("Array original:")

# analisaremos f (n) e g(n) para 0 n 4.

# algoritmo A:
f = n**2 + 2*n + 1
# algoritmo B:
g = 2*n

plt.figure(figsize=(10, 6))

# ******* Primeiramente, analisaremos f (n) e g(n) para 0 n 4 *********
# Plotar as funções
plt.plot(n, f, 'r-', label='f(n) = n^2 + 2n + 1')
plt.plot(n, g, 'b-', label='g(n) = 2n')

# Adicionar elementos ao gráfico
plt.title('Comparação das Funções f(n) e g(n)')
plt.xlabel('n')
plt.ylabel('Valores das Funções')
plt.grid(True)
plt.legend()

# mostrar o gráfico
plt.show
```
