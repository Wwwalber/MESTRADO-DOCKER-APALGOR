```python
import numpy as np
import matplotlib.pyplot as plt

# Teste simples de gráfico
x = np.linspace(0, 10, 100) #um array de 100 valores igualmente espaçados entre 0 e 10.
y = np.sin(x) # uma função da biblioteca NumPy que calcula o seno de cada valor no array x

plt.plot(x, y)
plt.title("Teste de Gráfico")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# algoritmo A
# f(n) = 10n

# algoritmo B
# f(n) = 2^n

n = np.arange(1, 20, 1) # array de 1 a 10, assim vou poder comparar os algoritmos para n de 1 a 10
algoritmo_A = 10*n**2
algoritmo_B = 2**n

plt.plot(n, algoritmo_A, label='Algoritmo A: 10n^2')
plt.plot(n, algoritmo_B, label='Algoritmo B: 2^n')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('Comparação de Algoritmos')
plt.yscale('log')  # Usando escala logarítmica para melhor visualização
plt.grid(True)
plt.legend()
plt.show()
```

```python
def soma_quadrados_A(n):
    x = 0
    for j in range(0, n):
        x = x + j * j
    return x

def soma_quadrados_B(n):
    x = n * (n + 1) * (2 * n + 1)
    x = x / 6
    return x

print(soma_quadrados_A(6)) # 385
print(soma_quadrados_B(6)) # 385
print(soma_quadrados_A(72)) # 338350
print(soma_quadrados_B(75)) # 338350
```
