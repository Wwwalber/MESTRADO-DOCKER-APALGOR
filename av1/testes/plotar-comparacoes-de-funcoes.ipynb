```python
import numpy as np
import matplotlib.pyplot as plt

# Definir o intervalo de valores para n
n = np.linspace(1, 100, 100)  # Cria 100 pontos entre 1 e 100

# Definir as funes
f = 0.5 * n**2 + 3 * n
g = n**2
h = 3*n
# Criar o grfico
plt.figure(figsize=(10, 6))

# Plotar as funes
plt.plot(n, f, 'r-', label='f(n) = 0.5n^2 + 3n')
plt.plot(n, g, 'b-', label='g(n) = n^2')
plt.plot(n, h, 'g-', label='h(n) = h = 3*n' )

# Adicionar elementos ao grfico
plt.title('Comparao das Funes f(n), g(n) e h(n)')
plt.xlabel('n')
plt.ylabel('Valores das Funes')
plt.grid(True)
plt.legend()

# Mostrar o grfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Definir o intervalo de valores para n
n = np.linspace(1, 100, 100)  # Cria 100 pontos entre 1 e 100

# Definir as funes
# Use np.log2 em vez de math.log
a = np.log2(n)
b = n
c = n * np.log2(n)
d = n**2

# Para o fatorial, vamos criar uma funo vetorizada
def factorial_array(n):
    return np.array([np.math.factorial(int(i)) for i in n])

e = factorial_array(n)

# Criar o grfico
plt.figure(figsize=(10, 6))

# Plotar as funes
plt.plot(n, a, 'r-', label='O(log n)')
plt.plot(n, b, 'b-', label='O(n)')
plt.plot(n, c, 'g-', label='O(n log n)')
plt.plot(n, d, 'y-', label='O(n^2)')
plt.plot(n, e, 'm-', label='O(n!)')

# Adicionar elementos ao grfico
plt.title('Comparao das Funes')
plt.xlabel('n')
plt.ylabel('Valores das Funes')
plt.grid(True)
plt.legend()

# Mostrar o grfico
plt.show()
```
