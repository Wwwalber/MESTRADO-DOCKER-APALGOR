```python
import sys
sys.path.append("../")  # Adiciona o diretrio raiz do projeto ao PATH

import numpy as np
import random
import matplotlib.pyplot as plt

def merge_sort(arr):
    # MERGE-SORT (A, p, r) // Na verso de pseudo cdigo p  o primeiro elemento e r  o ltimo
    if len(arr) > 1:
        meio = len(arr) // 2 # Encontra o meio do array
        L = arr[:meio] # Divide o array em duas metades
        R = arr[meio:]

        merge_sort(L) # Ordena a primeira metade
        merge_sort(R) # Ordena a segunda metade

        i = j = k = 0 # Inicializa os ndices

        while i < len(L) and j < len(R): # Copia os dados para os arrays temporrios L[] e R[]
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
            

                
#gerar dados de entrada
def gerar_dados_mergesort():
    n = np.arange(1, 501, 10)
    f_de_n_aleatorio = []
    f_de_n_crescente = []
    f_de_n_decrescente = []

    for t in n:
        array_aleatorio = [random.randint(1, 100) for _ in range(t)]
        lista_crescente = sorted(array_aleatorio)
        lista_decrescente = sorted(array_aleatorio, reverse=True)

        a_aleatorio = merge_sort(array_aleatorio, t)
        a_crescente = merge_sort(lista_crescente, t)
        a_decrescente = merge_sort(lista_decrescente, t)

        f_de_n_aleatorio.append(a_aleatorio)
        f_de_n_crescente.append(a_crescente)
        f_de_n_decrescente.append(a_decrescente)

    return n, f_de_n_aleatorio, f_de_n_crescente, f_de_n_decrescente

#plotar grfico
def plotar_merge_sort_comparativo(n, f_de_n_aleatorio, f_de_n_crescente, f_de_n_decrescente, titulo="MERGEESORT - Comparao de Dados de Ordenao"):
    plt.figure(figsize=(10, 6))
    plt.plot(n, f_de_n_aleatorio, label='Dados iniciais Aleatrios', color='blue')
    plt.plot(n, f_de_n_crescente, label='Dados iniciais Crescentes', color='red')
    plt.plot(n, f_de_n_decrescente, label='Dados iniciaisS Decrescentes', color='green')
    plt.xlabel('Tamanho do vetor (n)')
    plt.ylabel('Tempo de execuo f(n)')
    plt.title(titulo)
    plt.legend()
    plt.grid(True)
    plt.show()

n, f_de_n_aleatorio, f_de_n_crescente, f_de_n_decrescente = gerar_dados_mergesort()  

plotar_bubble_sort_comparativo(n, f_de_n_aleatorio, f_de_n_crescente, f_de_n_decrescente)
```
